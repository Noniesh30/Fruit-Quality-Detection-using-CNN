```python
#apply tranformations to avoid overfitting
#image augmentation
#it also helps in increasing size of data set
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

training_set = train_datagen.flow_from_directory(
    '/content/Drive/MyDrive/Colab Notebooks/aproject/dataset/training set',
    target_size=(64,64),
    batch_size=32,
    class_mode='categorical'
)
```

```python
test_datagen = ImageDataGenerator(
    rescale=1./255
)

testing_set = test_datagen.flow_from_directory(
    '/content/Drive/MyDrive/Colab Notebooks/aproject/dataset/test_data',
    target_size=(64,64),
    batch_size=32,
    class_mode='categorical'
)
```

```python
cnn = tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu',input_shape=[64,64,3]))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
#flattening
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=128,activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=10,activation='softmax'))
#softmax
```

```python
cnn.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
cnn.fit(x=training_set,validation_data=testing_set,epochs=25)
```
